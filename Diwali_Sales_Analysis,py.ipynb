```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head(10)
```

```python
df.drop(['Status','unnamed1'], axis = 1 , inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int32')
```

```python
df.dtypes['Amount']
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

### **Exploratory Data Analysis**

**Gender**

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df, palette = 'rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen
```

```python
ax = sns.barplot(x = 'Gender', y ='Amount', data = sales_gen, palette = 'rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

**From above graph we can see that most of the buyers if of female category and even the purchasing power of female is greater than men.**

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', data = df , hue = 'Gender', palette = 'rocket')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index= False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)

```

```python
sales_age
```

```python
ax = sns.barplot(data = sales_age, x = 'Age Group', y = 'Amount', palette = 'rocket')
```
